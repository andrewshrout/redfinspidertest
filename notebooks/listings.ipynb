```python
import requests
from parsel import Selector


headers = {
    'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:86.0) Gecko/20100101 Firefox/86.0'
}
response = requests.get('https://www.redfin.com/city/1823/AL/Birmingham', headers=headers)
selector = Selector(response.text)
```

```python
listing_links = selector.xpath('//a[@data-rf-test-name="basic-card-photo"]/@href').getall()
#print(listing_links)
for listing in listing_links:
    print(listing)
```

```python
# pagination / listings csv link
listings_csv = selector.xpath('//a[@id="download-and-save"]/@href').get()
print(listings_csv)
```
